```python
import os
os.chdir("/content/drive/MyDrive/python/moviepy/temp/mpeg")
import fitz
import ffmpeg
```

```python
import glob  #將案分組
```

```python
pdfpath = glob.glob("*.pdf")
print(pdfpath[0])
```

```python
from PyPDF2 import PdfFileReader
pdf = PdfFileReader(open(pdfpath[0],'rb'))  #("G:\我的雲端硬碟\\python\\moviepy\\temp\\mpeg\\test.pdf",'rb')
countpdf = pdf.getNumPages()
```

```python
import re
```

```python
for i in range(0,countpdf):   #PDF轉JPG
  pdffile = pdfpath[0]
  doc = fitz.open(pdffile)
  page = doc.load_page(i)  # number of page
  pix = page.get_pixmap()
  output = "test"+str(i+1)+".jpg"
  pix.save(output)
```

```python
m4a = glob.glob("*.m4a") #根據副檔名排序
m4a.sort()
jpg = glob.glob("*.jpg")
jpg.sort()
```

```python
print(jpg)
```

```python
for i in range(len(jpg)):
    
    input_still = ffmpeg.input(jpg[i])
    input_audio = ffmpeg.input(m4a[i]).filter('volume', 2)
    
    (
        ffmpeg
        .concat(input_still, input_audio, v=1, a=1)
        .output("temp"+str(i)+".mp4", pix_fmt='yuv420p')
        .run(overwrite_output=True)
    )
        
    with open("temp.txt", "a") as file:
        file.write( "file "+"temp"+str(i)+".mp4\n")
```
